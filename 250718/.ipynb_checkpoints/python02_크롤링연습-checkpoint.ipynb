```python
import requests, openpyxl #데이터 불러오기 requests, excel작업을 한다 openpyxl lib 가져오기
from bs4 import BeautifulSoup
from openpyxl.styles import Alignment, Font #엑셀 스타일 중 폰트 변경하기 위해 추출한 기능

wrap_alignment = Alignment(wrap_text=True)
bold_font = Font(bold=True)
header_font = Font(bold=True, size=14)
normal_font = Font(bold=False)

excel_file = openpyxl.Workbook() #excel_file이라는 변수는 인스턴스 객체가 됨
excel_sheet = excel_file.active #오픈된 신규 엑셀 파일 sheet안에 값을 저장하기 위한 준비
excel_sheet.title = "게시글" #sheet이름을 게시글로 해라

header = ["No", "게시글 제목", "댓글 수"] #행안에 들어가는 데이터 지정하기
excel_sheet.append(header)

for cell in excel_sheet[1]:
    cell.font = header_font
    
excel_sheet.column_dimensions["A"].width="20"#기준이 되어지는 칼럼의 넓이를 조정해라
excel_sheet.column_dimensions["B"].width="80"#기준이 되어지는 칼럼의 넓이를 조정해라
excel_sheet.column_dimensions["C"].width="80"#기준이 되어지는 칼럼의 넓이를 조정해라


res = requests.get("https://davelee-fun.github.io/trial/board/news.html") 
soup = BeautifulSoup(res.content, "html.parser")


num = 0
items = soup.select("div.list_item")
index = 0
row = 2


for item in items:
    title = item.select_one("span.subject_fixed")
    reply_count = item.select_one("span.rSymph05")
    links = item.select_one("a.list_subject")
    if title != None:
        num+=1
        #print(num, title.get_text().strip(), [int(replies_num.get_text().strip())])
        links_url = "https://davelee-fun.github.io/trial/board/"+links["href"]
        data1 = [num, title.get_text().strip(), "["+reply_count.get_text().strip()+"]"]
        excel_sheet.append(data1)
        excel_sheet.cell(row=row, column=2).hyperlink = links_url
        for cell in excel_sheet[row]:
            cell.font = bold_font
        row +=1
        res_title = requests.get("https://davelee-fun.github.io/trial/board/"+links["href"])
        title_soup = BeautifulSoup(res_title.content, "html.parser")
        replies = title_soup.select("div.comment_content")
        for reply in replies:
            data2 = ["","",reply.get_text().replace("\n","").replace("","\t").strip()]
            #print("ㄴ", reply.get_text().replace("\n","").strip())
            excel_sheet.append(data2)
            for cell in excel_sheet[row]:
                cell.font = normal_font
                cell.alignment = wrap_alignment #셀의 길이에 맞춰 정렬한다
            row+=1

excel_file.save("mysite.xlsx")
excel_file.close()
print("mysite.xlsx 저장이 완료되었습니다.")
```

```python
import requests, openpyxl #데이터 불러오기 requests, excel작업을 한다 openpyxl lib 가져오기
from bs4 import BeautifulSoup
from openpyxl.styles import Alignment, Font, colors #엑셀 스타일 중 폰트 변경하기 위해 추출한 기능

wrap_alignment = Alignment(wrap_text=True)
center_alignment = Alignment(horizontal="center", vertical="center") #수평, 수직 가운데 정렬, 동렬
bold_font = Font(bold=True)
header_font = Font(bold=True, size=14)
normal_font = Font(bold=False)
hyperlink_font = Font(color=colors.BLUE, underline="single")


excel_file = openpyxl.Workbook() #excel_file이라는 변수는 인스턴스 객체가 됨
excel_sheet = excel_file.active #오픈된 신규 엑셀 파일 sheet안에 값을 저장하기 위한 준비
excel_sheet.title = "게시글" #sheet이름을 게시글로 해라

header = ["No", "게시글 제목", "댓글 수"] #행안에 들어가는 데이터 지정하기
excel_sheet.append(header)

for cell in excel_sheet[1]:
    cell.font = header_font
    cell.alignment = center_alignment
    
excel_sheet.column_dimensions["A"].width="20"#기준이 되어지는 칼럼의 넓이를 조정해라
excel_sheet.column_dimensions["B"].width="80"#기준이 되어지는 칼럼의 넓이를 조정해라
excel_sheet.column_dimensions["C"].width="80"#기준이 되어지는 칼럼의 넓이를 조정해라


res = requests.get("https://davelee-fun.github.io/trial/board/news.html") 
soup = BeautifulSoup(res.content, "html.parser")


num = 0
items = soup.select("div.list_item")
index = 0
row = 2


for item in items:
    title = item.select_one("span.subject_fixed")
    reply_count = item.select_one("span.rSymph05")
    links = item.select_one("a.list_subject")
    if title != None:
        num+=1
        #print(num, title.get_text().strip(), [int(replies_num.get_text().strip())])
        links_url = "https://davelee-fun.github.io/trial/board/"+links["href"]
        data1 = [num, title.get_text().strip(), "["+reply_count.get_text().strip()+"]"]
        excel_sheet.append(data1)
        excel_sheet.cell(row=row, column=2).hyperlink = links_url
        excel_sheet.cell(row=row, column=2).font = hyperlink_font
        
        for cell in excel_sheet[row]:
            if cell.column != 2:
                cell.font = bold_font
            
        row +=1
        res_title = requests.get("https://davelee-fun.github.io/trial/board/"+links["href"])
        title_soup = BeautifulSoup(res_title.content, "html.parser")
        replies = title_soup.select("div.comment_content")
        for reply in replies:
            data2 = ["","",reply.get_text().replace("\n","").replace("","\t").strip()]
            #print("ㄴ", reply.get_text().replace("\n","").strip())
            excel_sheet.append(data2)
            for cell in excel_sheet[row]:
                cell.font = normal_font
                cell.alignment = wrap_alignment #셀의 길이에 맞춰 정렬한다
            row+=1

excel_file.save("mysite.xlsx")
excel_file.close()
print("mysite.xlsx 저장이 완료되었습니다.")
```
