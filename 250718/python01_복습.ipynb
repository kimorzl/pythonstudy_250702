```python
import openpyxl #엑셀 작업을 할 수 있게 해주는 lib
import re #정규표현식을 사용할 수 있는 lib

work_book = openpyxl.load_workbook("data_kr.xlsx") #openpyxl에 있는 load_workbook이라는 함수를 사용해 파일안에 있는 데이터를 불러온다.
work_sheet = work_book.active #work_sheet라는 변수에 work_book에서 오픈한 파일의 첫번째 시트 데이터를 불러온다

for each_row in work_sheet.rows:#sheet에 있는 각각의 행에 해당하는 데이터를 확인하다.
        #print(each_row[1].value)#각각의 요소 가운데 1번째 열에 속해있는 데이터를 출력해줘
        print(f"{each_row[0].value}님의 주민등록번호는 :", re.sub("-[0-9]{7}","-*******",each_row[1].value)) #re.sub은 regular extensible subtitute의 약자로 리스트의 replace와 유사한 역할을 한다. 
                                                        #단, sub활용시 ( 패턴, 바꾸고 싶은 형식, 바꿔야할 원본 데이터 정보)의 양식으로 작성해야함
                                                        # -[0-9]{7}의 의미 -다음 0~9까지의 숫자 7개가 반복되는 것
```

#### 크롤링 증권 사이트 -> 회사명만 크롤링 후 출력

```python
#크롤링 데이터 구조 파악하기
# 사이트 정보 불러오기
# 불러온 사이트에서 원하는 데이터 요소 불러오기
# 출력하기

#내가한 풀이
import requests
from bs4 import BeautifulSoup

num = 0
res = requests.get("http://davelee-fun.github.io/blog/crawling_stock_example.html") #url 사이트에서 데이터 불러오기
#print(res) response의 값이 200일때 성공적으로 데이터를 가져온 것이니 미리 확인 필수!

soup = BeautifulSoup(res.content, "html.parser") #불러온 파일에 해당하는 데이터 콘텐츠들을 html파서를 이용해 보기 쉽게 정렬하기, html이 아닌 xml형식이면 파싱할때 lxml로 파싱한다

#내가 데이터 찾은 코드
items = soup.select("li span.name") #selec를 이용해 불러온 데이터는 list 데이터 형태로 불러온다

for item in items: #items라는 범위 내에서 반복문을 실행해
    num+=1
    print("회사명", num, ":", item.get_text().replace(" ", "").replace("\n","")) #text값만 출력해라
```

#### 현재 가져온 데이터에서 회사명, 주식금액, 변동률을 같이 출력하라

```python
#크롤링 데이터 구조 파악하기
# 사이트 정보 불러오기
# 불러온 사이트에서 원하는 데이터 요소 불러오기
# 출력하기

#내가한 풀이
import requests
from bs4 import BeautifulSoup

num = 0
res = requests.get("http://davelee-fun.github.io/blog/crawling_stock_example.html") #url 사이트에서 데이터 불러오기
#print(res) response의 값이 200일때 성공적으로 데이터를 가져온 것이니 미리 확인 필수!

soup = BeautifulSoup(res.content, "html.parser") #불러온 파일에 해당하는 데이터 콘텐츠들을 html파서를 이용해 보기 쉽게 정렬하기, html이 아닌 xml형식이면 파싱할때 lxml로 파싱한다

#내가 데이터 찾은 코드
items = soup.select("li.row_sty") #selec를 이용해 불러온 데이터는 list 데이터 형태로 불러온다


for item in items: #items라는 범위 내에서 반복문을 실행해
    num+=1
    print(num,".",item.get_text().replace(" ", "").replace("\n"," ")) #text값만 출력해라
```

```python
#선생님풀이
import requests
from bs4 import BeautifulSoup

num = 0
res = requests.get("http://davelee-fun.github.io/blog/crawling_stock_example.html") 

soup = BeautifulSoup(res.content, "html.parser")

datas = soup.select("ul.table-stock.sty1>li.row_sty")
for data in datas:
    num+=1
    company = data.select_one("div.st_name").get_text().strip().replace(" ", "").replace("\n", "")
    price = data.select_one("div.st_price").get_text().strip().replace(" ", "").replace("\n", "")
    rate = data.select_one("div.st_rate").get_text().strip().replace(" ", "").replace("\n", "")
    print(num,".", f"{company}의 주가는{price}이고, 등락률은 {rate}입니다.")
```
