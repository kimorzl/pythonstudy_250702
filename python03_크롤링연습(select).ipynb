#### 2. 현재 쇼핑몰 내 우측 상단 Blog, About 메뉴명을 크롤링 후, 출력해 주세요.

```python
import requests
from bs4 import BeautifulSoup

url = "https://davelee-fun.github.io/"
res = requests.get(url)
soup = BeautifulSoup(res.content, "html.parser")

items = soup.select("li.nav-item")

for item in items:
    print(item.get_text())
```

#### 4. 현재 사이트에서 sitetitle이라는 클래스 값을 가지고 있는 h1 태그안 텍스트를 출력해주세요.

```python
import requests
from bs4 import BeautifulSoup

url = "https://davelee-fun.github.io/"
res = requests.get(url)
soup = BeautifulSoup(res.content, "html.parser")

items = soup.select_one(".sitetitle")
item = soup.select("h1.sitetitle")

for i in item :
    print(i.get_text())
    
print(items.get_text())
print(items.string)
```

#### 5. 현재 사이트에서 "선물하기 좋은 선물을 고르고 골라서 소개하는 블로그입니다."라는 서브타이틀을 찾아 크롤링 후, 출력해주세요.

#### 6.  현재 사이트에서 featured 카테고리에 속해있는 상품들의 상품명만 가져와 출력해주세요.

```python
import requests
from bs4 import BeautifulSoup

url = "https://davelee-fun.github.io/"
res = requests.get(url)
soup = BeautifulSoup(res.content, "html.parser")

items = soup.select("h4.card-text")

for item in items :
    print(item.get_text().split(":")[1].replace("\n",""))
```

#### 7. 현재 사이트에서 All stories 카테고리에 속해있는 상품들의 상품명만 가져와 출력해주세요.

```python
#내 풀이
import requests
from bs4 import BeautifulSoup

url = "https://davelee-fun.github.io/"
res = requests.get(url)
soup = BeautifulSoup(res.content, "html.parser")

allstories = soup.select(".recent-posts .card-text")
#print(allstories)


for item in allstories :
    print(item.get_text().split(":")[1].replace("\n",""))
```

```python
#선생님 풀이
import requests
from bs4 import BeautifulSoup

url = "https://davelee-fun.github.io/"
res = requests.get(url)
soup = BeautifulSoup(res.content, "html.parser")

allstories = soup.select_one(".recent-posts")
items = allstories.select("h4.card-text")
#print(allstories)


for item in items :
    print(item.string.strip().split(":")[1])
```
