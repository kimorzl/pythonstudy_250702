```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By 
from webdriver_manager.chrome import ChromeDriverManager 
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import time 
import pyautogui

service = Service(ChromeDriverManager().install())

options = Options()
options.add_argument("headless") #사이트에서 막는 경우도 있음
options.add_argument("window-size=1920x1080")
options.add_argument("--disable-gpu") 
options.add_argument("--User-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/138.0.0.0 Safari/537.36")
options.add_argument("lang=ko_KR") #우리가 만든 문서는 한국어다

driver = webdriver.Chrome(service=service, options=options)
driver.get("https://davelee-fun.github.io/")

for i in range(5):
    print(driver.current_url) #내가 지금 보고있는 url에 대한 페이지를 가져와라
    elements = driver.find_elements(By.CSS_SELECTOR, "h4.card-text")
    for element in elements :
        print(element.text.split(':')[1])
    
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight)") #JAVA의 문법을 사용하겠다 excute_script
        #document.body.scrollHeight : 스크롤 할 수 있는 높이의 끝까지 해라
    
    elements = driver.find_elements(By.CSS_SELECTOR, ".ml-1.mr-1")
    elements[-1].click()

    time.sleep(3)

driver.quit()
```
