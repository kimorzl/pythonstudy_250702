- scrapy
    * 크롤링 전문 프레임워크
    * 현업에서 대량의 데이터 크롤링이 필요할 경우, 활용함
    * 디버깅에 강해 서버가 안정적이며 오류를 명확하게 알려줌
    * 장점
      * 속도 reqeuests < selenium < scrapy
      * 속도가 빠른 이유는 비동기처리방식 (병렬구조)를 가지고 있기 때문이다
      * 단, 터미널을 통해 명령어를 자주 입력해야한다.
      * 터미널에서 명령어 입력시 cli환경이 되어야하는데 그러기 위해선 vs code 설치가 전제조건이 됨
          * 아나콘다, 주피터노트북은 컴파일러라고 하며 웹 에디터인 vs code와 연동되어야함
- scrapy 사용방법

  * crapy startproject(프로젝트 명) /scrapy startproject = 예약어
  * scrapy.cfg : 배포용 시스템 설정파일(크롤링 이후 수집해온 데이터를 어떻게 서버에 배포할 것인지)
  * python -m pip install scrapy : 우회설치
  * helloscrapy : 프로젝트 폴더명
  * __init__.py : 현재 프로젝트 폴더를 파이썬 패키지로 만들어주는 초기역할, 스크래피도 기존 파이썬 문법처럼 이미 누군가가 만들어놓은 라이브러리 활용 가능
  * items.py : 프로젝트 내 아이템 정의 파일, 크롤링을 해오는데 예를 들어 A데이터 찾아오기 / B데이터 찾아오기/ C데이터 찾아오기 => 크롤링을 하기 위한 각각의 실행 명령 아이템을 사전에 정의해놓는 파일
  * pipelines.py : 앞에서 정의해 놓은 실행 명령 아이템들을 어떤 순서 및 방법으로 처리할 것인지 설정 해놓은 파일
  * settings.py : 프로젝트 배포를 위한 실제 설정 파일. 스크래피를 통해서 크롤링 시, 동시에 실행할 크롤링 수는 몇개, 미들웨어는 어떤것을 사용할지 등과 같은 기타 세부적인 설정을 하는 파일
  * spiders : 실제 크롤링을 하는 역할 > 크롤러(*스파이더) > 외부 라이브러리를 가져다가 사용할 수 있게끔 하기위해 __init__.py가 존재함
  * middlewears.py : 비동기적으로 실행되는 프로그래밍 언어들에 존재하는 개념적인 요소로 주로 최종 데이터 집계 전 각각의 아이템들의 정보를 수집한 후 최종 연산처리에 사용됨

실제 크롤링을 담당하는 크롤러(*=스파이더)는 spiders>orzl.py라는 파일 안에 class로 정의!!!
이때, class는 scrapy.Spider(*스크래피의 창조주 = 개발자)라는 기본값을 상속받아서 생성됨

기본 셋팅값
1. name = 크롤러의 이름
2. allowed_domains : 크롤링해올 사이트의 도메인 리스트
3. start_urls : 크롤링을 시작할 메인 URL 리스트를 지정
4. 크롤러에게 본격적으로 크롤링해야할 대상, 방법 설정 - 함수를 활용함
    * response
      - response.css('selector::text').get() : css선택자를 사용해 텍스트를 추출하는 방식 **참고**selenium크롤링의 find_element와 유사한 성격
      - response.xpath('xpath').get() : xpath를 사용해 요소를 선택하고 텍스트를 추출하는 방식
      - get() : 선택한 요소의 첫 번째 결과를 찾으면 해당 결과값을 반환 후 종료/찾아올 결과값이 없으면 none을 반환
      - getall() : 선택한 모든 요소 반환, 만약 값이 없다면 빈 '리스트' 반환하라

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By 
from webdriver_manager.chrome import ChromeDriverManager 
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
from urllib.request import urlretrieve #url에서 요청한 자료를 수집해오는 urllib내 인스턴스객체
import time 
import pyautogui

service = Service(ChromeDriverManager().install())

options = Options()
options.add_argument("headless") 
options.add_argument("window-size=1920x1080")
options.add_argument("--disable-gpu") 
options.add_argument("--User-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/138.0.0.0 Safari/537.36")
options.add_argument("lang=ko_KR")

driver = webdriver.Chrome(service=service, options=options)
driver.get("https://davelee-fun.github.io/")

elements = driver.find_elements(By.CSS_SELECTOR, "div.maxthumb img")
sources = list()


#사이트에서 긁어온 src요소를 리스트 형태의 자료형으로 붙여넣기
for element in elements:
    sources.append(element.get_attribute("src"))

#인덱스 값을 부여해 각 값을 저장하는 법)
for index, source in enumerate(sources) :
    urlretrieve(source, f"IMG/{str(index)}.{source.split(".")[-1]}")


time.sleep(3)

driver.quit()
```

#### HTML 크롤링을 위한 XPATH 기본 포맷 (*형태)
- //tagname[@Attribute="vaule"]
- / : 절대 경로를 의미  (ex. /html/body/div/a)
- // : (상대 경로) 문서 내에서 바로 검색을 하라는 의미 (ex. //h1 => 문서에서 바로 h1태그를 찾아와라)
- //*[@href] : 문서 내, href라는 속성을 가지고 있는 모든 태그를 다 찾아와라
- //a[@href="https:///www.google.com"] -> 문서 내 a태그 중 href 속성값으로 구글 닷컴을 찾아와라
- (//a)[3] : 문서 내 a태그를 찾아오는데, 세번째 a만 찾아와라 (주의사항, 일반적 인덱스 값과 다르게 1번부터 찾아옴) // ()는 우선 연산자를 의미
- (//table)[last()] : 문서 내 table태그 중 마지막번째 table만 찾아와라
- (//a)[position() < 3] : 문서 내 a태그를 찾아와, 1번째 2번째 a만 찾아와라
- //div[@*] #문서내 div태그 중, 어떤 속성이든 상관없이 속성이 하나라도 있으면 찾아와라

```python
#XPATH활용
# ID : erroro@error.com 입력
# 비밀번호 : 1234 입력
# 로그인 버튼 클릭
# 로그인 페이지에서 message 클래스 속성값의 텍스트 출력
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By 
from webdriver_manager.chrome import ChromeDriverManager 
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
from urllib.request import urlretrieve
import time 
import pyautogui

service = Service(ChromeDriverManager().install())

options = Options()
#options.add_argument("headless") 
options.add_argument("window-size=1920x1080")
options.add_argument("--start-maximized")
options.add_argument("--disable-gpu") 
options.add_argument("--User-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/138.0.0.0 Safari/537.36")
options.add_argument("lang=ko_KR")

driver = webdriver.Chrome(service=service, options=options)
driver.get("https://divjason.github.io/sellenium-test/index.html?")

# input태그 id값으로 username이 들어가는 요소 찾기
element = driver.find_element(By.XPATH, "//input[@id='username']")
#값이 있다면 비워내기
element.clear()
#비워진 자리에 error@error.com 넣기
element.send_keys("error@error.com")
# input태그 id값으로 password 들어가는 요소 찾기
element = driver.find_element(By.XPATH, "//input[@id='password']")
#값이 있다면 비워내기
element.clear()
#비워낸 자리에 1234값 입력하기
element.send_keys("1234")

#3초간 기다리기
time.sleep(3)

#input태그 내 type속성값으로 submit을 가지고 있는 요소 찾기
element = driver.find_element(By.XPATH, "//input[@type='submit']")
#해당 요소 클릭하기 (버튼형이라는 소리)
element.click()
#div라는 태그 내 클래스 값으로 message를 갖는 요소 찾기
element = driver.find_element(By.XPATH, "//div[@class='message']")
#출력하기, 만일 element(단일요소)가 아닌경우 for in문을 이용해 elements(복수요소)로 출력하기
print(element.text)

driver.quit()
```
