```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import time #크롤링 파이썬 시간과 selenium을 통해 크롤링하는 데이터 지연시간 발생으로 인한 오차, 오류가 발생하는것을 막는 라이브러리
from openpyxl import Workbook



# 1. 셀레니움 기본 옵션 설정
options = Options()
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
options.add_argument("--headless")

driver = webdriver.Chrome(options = options)

# 2. 페이지 접속
url = "https://www.kurly.com/goods/1000202976"
driver.get(url)

time.sleep(5) #페이지 로딩 대기 시간

# 3. HTML 가져오기
soup = BeautifulSoup(driver.page_source, "html.parser")

# 4. review 텍스트 추출하기
reviews = soup.select("p.css-y49dcn.e36z05c13")
review_lists = [review.get_text().strip() for review in reviews]

driver.quit() #selenium 크롤링으로 찾아온 데이터가 있다면 꼭 끝내줘야한다.

wb = Workbook()
ws = wb.active
ws.title = "Kurly Reviews"

ws.append(["리뷰 내용"])
for item in review_lists :
    ws.append([item])

wb.save("kurly_reviews.xlsl")
print("엑셀 저장 완료 : kurly_reviews.xlsx")
```
